```python
import torch
import pandas as pd
from timesfm import TimesFm, TimesFmCheckpoint, TimesFmHparams
from foundation_models_for_finance.timesfm_loader import load_timesfm_from_hf
```

```python
repo_id = "FinText/TimesFM_8M_2023_Augmented"
num_samples = 20
prediction_length = 1
window_size = 21
tfm = load_timesfm_from_hf(repo_id=repo_id, model_size="8M", prediction_length=prediction_length)
```

```python
df_excess_ret = pd.read_csv('../data/two_stocks_excess_returns.csv', index_col=0, parse_dates=True)
input = torch.tensor(df_excess_ret.iloc[:window_size].values.T)
input.shape
```

```python
point_forecast, quantile_forecast = tfm.forecast(input)
```

```python
# the point forecast is the mean prediction
print("Point Forecast shape:")
print(point_forecast.shape)
# the quantile forecast contains predictions for mean, 10%, 20%, ..., 90% quantiles
print("Quantile Forecast shape:")
print(quantile_forecast.shape)  # (num_series, prediction_length, num_quantiles)
```
